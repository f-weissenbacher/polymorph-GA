```python
import imolecule

from IPython.display import display, HTML
carbons = ['CCCC'] * 16
renders = (imolecule.draw(c, size=(150, 150), shader='lambert', display_html=False) for c in carbons)
```

```python
columns = ('<div class="col-xs-6 col-sm-3">{}</div>'.format(r) for r in renders)
display(HTML('<div class="row">{}</div>'.format("".join(columns))))
```
